Creating new dataset


```python
import pandas as pd
```

```python
df=pd.read_csv('new_df.csv')
df
```

```python
df1=pd.read_csv('dataset.csv')
df1
```

```python
def remove_space_between_word(dataset):
    for col in dataset.columns:
        for i in range(len(dataset[col])):
            if (type(dataset[col][i]) == str ):
                dataset[col][i] = dataset[col][i].strip()
                dataset[col][i] = dataset[col][i].replace(" ", "_")
    return dataset
```

```python
df1=remove_space_between_word(df1)
df1.head()
```

```python
df1=df1.fillna(0)
```

```python
def create_new_df(df1,df2):
    for i in range(4920):
        for col in df2.columns:
            if col=='Disease':
                continue
            df1.at[i,[df2[col][i]]]=1
    return df1
```

```python
n_df=create_new_df(df,df1)
```
